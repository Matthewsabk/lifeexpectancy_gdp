## Life Expectancy and GDP
#### Goal: Analyze a provided dataset to attempt to find relationships between GDP and life expectancies in six countries. 

The dataset includes 96 entries for four data fields: Country, Year, Life expectancy at birth in years, and GDP. It spans the time period of 2000 to 2015 for six countries: Chile, China, Germany, Mexico, The United States of America, and Zimbabwe.

It was provided by Codecademy as part of their Data Science coursework and was originally pulled from the World Health Organization and the World Bank.

```python
#libraries
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
#set display options: suppress scientific notation
pd.options.display.float_format = '{:,.0f}'.format
```

```python
#import data
df = pd.read_csv('all_data.csv')
#rename columns for python use
df.rename(columns={'Life expectancy at birth (years)': 'life_expectancy', 'Country':'country', 'Year':'year', 'GDP':'gdp' }, inplace=True)
#replace Unites States of America with USA in country column
df['country'].replace({'United States of America': 'USA'
}, inplace=True)
```

#### What is the relationship of GDP and Life Expectancy across the data set?

```python
#set figure for Life Expectancy vs GDP plots
fig, ax = plt.subplots(2,3,figsize=(10,8))
#set the initial plot position
pos = 1
#for loop to itterate through countries in the dataset
for country in list(df['country'].unique()):
    country_df = df[df['country']==country]
    #create a subplot for each country
    plt.subplot(2,3,pos)
    #plot the lineplot of life expectancy vs gdp, supress the legend
    ax = sns.lineplot(y='life_expectancy', x='gdp', data=country_df, label = 'Life Expectancy vs GDP - literals', legend=None)
    #plot the regression plot of life expectancy vs gdp
    regres = sns.regplot(y='life_expectancy', x='gdp', data=country_df, order=2, label = "Polynomial Regression Best Fit")
    #after plotting the data, format the labels
    xlabels = [str(x) + 'T' for x in ax.get_xticks()/1000000000000]
    plt.gca().set_xticks([x for x in ax.get_xticks()])
    plt.gca().set_xticklabels(xlabels, rotation = 90)
    ax.set(xlabel=None, ylabel=None)
    #title each plot
    plt.title(str(country).title())
    #increment position of next plot
    pos +=1
#label and format the figure
fig.suptitle('Life Expectancy vs GDP by Country')
fig.supxlabel('GDP in Trillions (USD)')
fig.supylabel('Life Expectancy (Years)')
plt.subplots_adjust(hspace=0.28, wspace=0.22)
handles, labels = plt.gca().get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')
#show and clear the figure
plt.show()
plt.close('all')
```

The relationship between GDP and Life Expectancy is easiest to see when the dataset is broken down by country. In each country there is a clear positive non-linear correlation between GDP and Life Expectancy. The steepest increases in Life Expectancy seen at lower GDP's.
